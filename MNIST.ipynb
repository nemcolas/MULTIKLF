```python
from sklearn.datasets import fetch_openml
```

```python
dataset = fetch_openml('mnist_784', version=1)
dataset
```

```python
X = dataset.data
y = dataset.target
print(X.shape, y.shape)
```

```python
X.sample(5)
```

```python
numero_aleatorio = X.sample()
display(numero_aleatorio)
```

```python
plt.imshow(numero_aleatorio.values.reshape(28, 28), cmap='gray_r')
plt.show()
```

```python
import numpy as np

np.sqrt(784)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
clf.predict(numero_aleatorio)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
print ("indice", "certo", "predito", sep = ' - ')
for i in range(len(y_test)):
    print(i, y_test.iloc[i], y_pred[i], sep = ' - ')

    if i == 10:
        break
```
